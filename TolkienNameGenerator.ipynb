```python
import pandas as pd
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Tolkien_characters.csv')
raw_names = data['Name'].tolist()
```

```python
names = []
for name in raw_names:
    name = name.lower()
    if ' ' in name:
        name = '0'.join(name.split())
    if '-' in name:
        name = '1'.join(name.split('-'))
    name = name + '2'
    names.append(name)

max_len = max((len(s) for s in raw_names))
total_len = sum((len(s) for s in raw_names))
different_char = set("".join(names))
total_char = len(different_char)


print(f'The largest name has {max_len} characters. ')
print(f'The total number of characters is {total_len}.')
print(f'The total number of different characters is {total_char}.')
print(sorted(different_char))
```

```python
tokenizer = tf.keras.preprocessing.text.Tokenizer(num_words=len(different_char))
for name in names:
    tokenizer.fit_on_texts(list(name))
sequences = tokenizer.texts_to_sequences(np.array([list(name) for name in names]))
```

```python
input_sequences = []
labels = []
for line in sequences:
    for i in range(1, len(line)):
        input_sequences.append(line[0:i])
        labels.append(line[i])
X = tf.keras.preprocessing.sequence.pad_sequences(input_sequences, max_len-1)
y = tf.keras.utils.to_categorical(labels, num_classes=total_char)
```

```python
num_dims = 16
num_units = 32
num_epochs = 200

model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Embedding(total_char, num_dims, input_length=max_len-1))
model.add(tf.keras.layers.LSTM(num_units))
model.add(tf.keras.layers.Dense(total_char, activation='softmax'))
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
history = model.fit(X, y, epochs=num_epochs, verbose=1)
model.save('model/model.h5')
```

```python
loss = history.history['loss']
accuracy = history.history['accuracy']

plt.figure(figsize=(12, 4))
plt.subplot(121)
plt.plot(range(0, num_epochs), loss, 'orange')
plt.ylim([0, 4])
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Loss over epochs', fontsize=12)
plt.grid(alpha=0.2)

plt.subplot(122)
plt.plot(range(0, num_epochs), accuracy)
plt.ylim([0, 1])
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.title('Accuracy over epochs', fontsize=12)
plt.grid(alpha=0.2)
plt.show()
```

```python
def build_name(out_name):
    name = []
    for c in out_name:
        if c == '0':
            name.append(' ')
        elif c == '1':
            name.append('-')
        elif c != ' ':
            name.append(c)
    name = ''.join(name)
    return ' '.join([n.capitalize() for n in name.split()])


def generate_name(input_name=''):
    while True:
        gen_name = input_name
        gen_name = gen_name.lower()
        gen_name = '0'.join(gen_name.split(' '))  # 0: Spaces
        gen_name = '1'.join(gen_name.split('-'))  # 1: Hyphens
        char = '' 
        while char != '2':  # 2: End of name
            token_list = tokenizer.texts_to_sequences([' '.join(gen_name)])
            token_list = tf.keras.preprocessing.sequence.pad_sequences(token_list, max_len-1)
            predicted = np.random.choice(range(0, total_char), p=model.predict(token_list)[0])
            for char, index in tokenizer.word_index.items():
                if index == predicted:
                    if char == '2':
                        break
                    gen_name += ' ' + char
        gen_name = build_name(gen_name)
        if gen_name not in raw_names:
            break
    return gen_name

new_name = generate_name()
print(new_name)
```
